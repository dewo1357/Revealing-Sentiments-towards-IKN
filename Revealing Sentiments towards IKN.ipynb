```python
data_kata=[]
except_alpha = ['a','i','u','e','o']
for i in range(len(df)):
    value = []
    for k in range(1,len(df['text'][i])):
        if df['text'][i][k-1] == df['text'][i][k]:
            if df['text'][i][k-1] not in except_alpha:
                value.append(df['text'][i][k])
                continue
            else:
                continue
        value.append(df['text'][i][k-1])
    data_kata.append(''.join(value))
df['text'] = data_kata
```

```python
def catch_kalimat(dataframe):
    kalimat = []
    text = []
    symbol = ['.',',']
    teks_data = []
    for baris in range(len(dataframe)):
        keseluruhan = []
        for i in dataframe['text'][baris]:
            if i == " " or i == "\n" or i in symbol:
                if i == " " or i in symbol:
                    pass
                else:
                    text.append(i)
                    keseluruhan.append(i)
                kalimat.append("".join(text))
                text = []
                continue
            text.append(i)
            
    for i in kalimat:
        if i == "" or len(i) < 4:
            kalimat.pop(kalimat.index(i))
    return kalimat
```

```python
positif = []
negatif = []
kemerdekaan = []
kepulauan = []

def konteks_kata(kalimat,title = ['Positif','kata Negatif','Semangat','Ibu Kota'], \
                 jenis_kalimat = [kata_positif,kata_negatif,kata_positif_kemerdekaan,
                                  ibu_kota],section = [positif,negatif,kemerdekaan,kepulauan]):
    for i in kalimat:
        for segment in range(len(jenis_kalimat)):
            for k in jenis_kalimat[segment]:
                if i == k or i == k.capitalize() or i == k.upper() or i == k.lower():
                    section[segment].append(i.capitalize()) 
    data = {
        'Konteks Kata' : [],
        'Jumlah' : []
    }
    for k in range(len(section)):
        data['Konteks Kata'].append(title[k])
        data['Jumlah'].append(len(section[k]))
    return pd.DataFrame(data)
    
data = konteks_kata(kalimat)
data.sort_values(by='Jumlah',ascending=False)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Konteks Kata',y='Jumlah',data=data.sort_values(by='Jumlah',ascending=False))
```

Kata-kata dengan ukuran lebih besar mengindikasikan tema atau isu yang lebih dominan dalam sentimen positif atau negatif. Hasil ini dapat membantu dalam memahami distribusi sentimen dalam data yang dianalisis, serta mengidentifikasi kata-kata kunci yang paling berpengaruh dalam membentuk persepsi positif atau negatif.

```python
point = []
positif = positif + kemerdekaan
for i in range(len(df)):
    point_p = 0
    point_n = 0
    point_k = 0
    for p in positif:
        if p.lower() in df['text'][i].lower():
            point_p +=1
    for m in negatif:
        if m.lower() in df['text'][i].lower():        
            point_n +=1
    for m in kepulauan:
        if m.lower() in df['text'][i].lower():
            point_k +=1
    if point_k > 0:
        point.append('Membahas IKN')
        continue
    if point_p == 0 and point_n == 0:
        point.append('Out Of Topic')
    elif point_p > point_n:
        point.append('Positif')
    elif point_p == 0 and point_n > 0:
        point.append('Negatif')
    else:
        point.append('no_detect')
df['kategori'] = point
```

Kode ini berfungsi untuk mengkategorikan teks berdasarkan sentimen (positif atau negatif) dan menentukan apakah teks tersebut relevan dengan topik IKN atau tidak. Hasil klasifikasi ini kemudian ditambahkan ke dalam DataFrame sebagai informasi tambahan.
